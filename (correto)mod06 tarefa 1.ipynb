####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(1, 2)
sns.countplot(ax = axes[0], x='APGAR1', data=data)
sns.countplot(ax = axes[1], x='APGAR5', data=data)
plt.subplots_adjust(wspace=0.3, hspace=0)
```

```python
ax = sns.countplot(x='GESTACAO', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right') #solução do seaborn para rotacionar e não sobrepor o rótulo X
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar5_consultas = data.groupby('CONSULTAS')['APGAR5'].mean().reset_index() #agrupando CONSULTAS e APGAR5, aplicando média e resetando o índice para CONSULTAS
```

```python
apgar5_consultas
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=apgar5_consultas)
plt.title('APGAR5 Médio por CONSULTAS')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.show()
```
